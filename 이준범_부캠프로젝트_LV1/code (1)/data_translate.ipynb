```python
import googletrans
import pandas as pd
import random

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
train_data = pd.read_csv('C:/Users/LG/부캠 LV1프로젝트/data/train.csv')
```

```python
    

sentences_1 = train_data['sentence_1'].to_list()
sentences_2 = train_data['sentence_2'].to_list()

new_df = train_data
new_df.tail()

index_lis = [i for i in range(len(train_data))]
for sentence_1, sentence_2, id in zip(sentences_1, sentences_2,index_lis):
    sentence_1 = sentence_1.split(' ')
    sentence_2 = sentence_2.split(' ')
    random.shuffle(sentence_1)
    random.shuffle(sentence_2)    
    new_data = {'id': id+len(train_data)+1,
               'source':  train_data.iloc[id]['source'],
               'sentence_1': ' '.join(sentence_1),
               'sentence_2': ' '.join(sentence_2),
                'label': train_data.iloc[id]['label'],
                'binary-label': train_data.iloc[id]['binary-label']
               }
    new_data = pd.DataFrame.from_dict([new_data])
    new_df = pd.concat([new_df, new_data], ignore_index=True)
    
new_df.tail()


```
